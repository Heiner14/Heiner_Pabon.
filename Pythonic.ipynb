###Ejercicio 1. Filtrar pares
Código base (no Pythonic):
numeros = [1, 2, 3, 4, 5, 6]
pares = []
for i in range(len(numeros)):
if numeros[i] % 2 == 0:
pares.append(numeros[i])
print(pares)
Objetivo: Usa una list comprehension.

```python
numeros = [1, 2, 3, 4, 5, 6]
pares = []
for i in range(len(numeros)):
if numeros[i] % 2 == 0:
pares.append(numeros[i])
print(pares)
```
